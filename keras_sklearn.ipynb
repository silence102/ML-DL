```python
# 코사인 유사도(방향성)
```

```python
import numpy as np
from numpy import dot
from numpy.linalg import norm
```

```python
# norm 
# 벡터 성분의 절대값으 모두 더하면 구할 수 있음.
# 벡터의 유클리드 거리로 구할 수 있음.
```

```python
def cos_sim(A, B):
    return dot(A, B)/(norm(A) * norm(B))
```

```python
"""
문서 1: 나는 치킨 좋아요
문서 2: 나는 피자 좋아요
문서 3: 나는 피자 좋아요 나는 피자 좋아요

        나는 치킨 피자 좋아요
문서 1:   1   1    0    1
문서 2:   1   0    1    1
문서 3:   2   0    2    2
"""
```

```python
doc1 = np.array([1,1,0,1])
doc2 = np.array([1,0,1,1])
doc3 = np.array([2,0,2,2])
```

```python
# 문서 1: 나는 치킨 좋아요
# 문서 2: 나는 피자 좋아요
# 문서 3: 나는 피자 좋아요 나는 피자 좋아요
print(cos_sim(doc1, doc2))
print(cos_sim(doc1, doc3))
print(cos_sim(doc2, doc3))
```

```python
# 코사인 유사도(두 벡터 간의 코사인 각도를 이용)
# 벡터 유사도로 방향이 완전히 동일한 경우 1 
# 90%의 각을 이루면 0
# 180도 반대 방향을 가지면 -1
```

```python
from sklearn.metrics.pairwise import cosine_similarity
print(cosine_similarity([doc1],[doc2]))
print(cosine_similarity([doc1],[doc3]))
print(cosine_similarity([doc2],[doc3]))
```

```python
# 유클리디안(유클리드 거리)
def eu_dist(A, B):
    return np.sqrt(np.sum((A-B)**2))
```

```python
# 문서 1: 나는 치킨 좋아요
# 문서 2: 나는 피자 좋아요
# 문서 3: 나는 피자 좋아요 나는 피자 좋아요
print(eu_dist(doc1, doc2))
print(eu_dist(doc1, doc3))
print(eu_dist(doc2, doc3))
```

```python
from sklearn.metrics.pairwise import euclidean_distances
print(euclidean_distances([doc1], [doc2]))
print(euclidean_distances([doc1], [doc3]))
print(euclidean_distances([doc2], [doc3]))
```

```python
data = df_movie[['title','overview']]
data
```

```python
data.info()
```

```python
data = data.dropna()
data.info()
```

```python
tfidf = TfidfVectorizer(stop_words="english")
tfidf_dtm = tfidf.fit_transform(data["overview"])
tfidf_dtm
```

```python
cos_sim_res = cosine_similarity(tfidf_dtm, tfidf_dtm)
```

```python
title_to_index = dict(zip(data["title"], data.index))
title_to_index
```

```python
def get_recommendation(title, n):
    idx = title_to_index[title]
    sim_scores = list(enumerate(cos_sim_res[idx]))
    sim_scores = sorted(sim_scores, key=lambda x : x[1], reverse=True)
    sim_scores_n = sim_scores[1:n+1]
    movie_idx = [ movie_dict[0] for movie_dict in sim_scores_n]
    return data["title"].iloc[movie_idx]
```

```python
get_recommendation("Toy Story", 10)
```
